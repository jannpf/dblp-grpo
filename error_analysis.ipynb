```python
import os
import json
import random
import dotenv
```

```python
dotenv.load_dotenv()
```

```python
EXPERIMENT = os.getenv("EXPERIMENT")
EXPERIMENT
```

```python
input_path = f"{EXPERIMENT}/results/20251129-124149_results.json.evaluation.json"
output_path = "error_sample.json"
```

```python
with open(input_path, 'rt') as f:
    results = json.load(f)
```

```python
incorrect_items = results["mismatches"] + results["errors"]
len(incorrect_items)
```

```python
sample = random.sample(incorrect_items, 100)
```

```python
for i in range(len(sample)):
    s = sample[i]
    for r in ("generated", "reference"):
        if r not in s:
            continue
        if len(s[r]) > 10:
            s[r] = s[r][:10]
            s[r].append("...")
    # for k in ("temporal", "held_out", "fbeta"):
    #     del s[k]
```

```python
with open(output_path, "wt") as f:
    json.dump(sample, f)
```
